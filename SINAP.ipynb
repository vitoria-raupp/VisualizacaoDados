```python
import pandas as pd
import numpy as np
```

```python
sinapi = pd.read_excel('Trabalho com Dados SINAPI.XLS')
sinapi.head()
```

```python
sinapi.columns
```

```python
sinapi.columns = list(sinapi.iloc[5])
sinapi
```

```python
sinapi.drop(range(0,6), inplace = True)
sinapi
```

```python
sinapi.reset_index(inplace = True)  # muda index
sinapi
```

```python
sinapi = sinapi[0:-2] # remove as duas ultimas (-)
sinapi.tail()
```

```python
ultimo = len(sinapi) -1
penultimo = len(sinapi) -2
sinapi.drop([ultimo, penultimo], inplace = True) 
sinapi.tail()
```

```python
# sinapi['PRECO MEDIANO R$'] = sinapi['PRECO MEDIANO R$'].astype('float') 
```

```python
def Ponto_nada(x):
  x = str(x).replace('.', '')
  return x
```

```python
def Virgula_ponto(x):
  x = str(x).replace(',', '.')
  return float(x)
```

```python
Ponto_nada("1.950,60")
```

```python
Virgula_ponto("1.950,60")
```

```python
sinapi['  PRECO MEDIANO R$'] = sinapi['  PRECO MEDIANO R$'].apply(Ponto_nada)
```

```python
sinapi['  PRECO MEDIANO R$'] = sinapi['  PRECO MEDIANO R$'].apply(Virgula_ponto)
```

```python
sinapi['  PRECO MEDIANO R$'] = sinapi['  PRECO MEDIANO R$'].astype('float')
```
