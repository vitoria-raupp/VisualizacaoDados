```python
import requests
import pandas as pd
```

```python
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
```

```python
resp = requests.get(url)
resp
```

```python
dados = resp.json()
dados
```

```python
dados[0]['title']
```

```python
issues = pd.DataFrame(dados, columns = ['number', 'title', 'labels', 'state'])
issues
```
