```python
import numpy as np
import pandas as pd
dados = {'Id': np.arange(0,10), 'Nascimento':  np.random.randint(low=10000000, high=100000000, size=10), 'CPF:': np.random.randint(low=10000000000, high=100000000000, size=10) , 'Nome': ['João', 'Maria', 'Pedro', 'Ana', 'Carlos', 'Laura', 'Fernando', 'Isabela', 'Rafael', 'Mariana']}

frame = pd.DataFrame(dados)
frame
```
