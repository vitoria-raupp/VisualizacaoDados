```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnarDataSource
import pandas as pd
```

```python
output_notebook()
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
```

```python
p = figure(title = 'Grafico de linha', x_axis_label = 'x', y_axis_label = 'y')
```

```python
p.line(x,y,legend_label = 'linha 1', line_width= 2)
```

```python
show(p)
```

```python
url = 'http://raw.githubusercontent.com/plotly/datasets/master/iris.csv'
```

```python
df = pd.read_csv(url)
```

```python
from bokeh.models.widgets.tables import ColumnDataSource
source = ColumnDataSource(df)
```

```python
u = figure(title = 'Grafico de dispers√£o', x_axis_label = 'Sepal Lengyh', y_axis_label = 'Petal Length')
```

```python
u.circle('SepalLength', 'PetalLength', source = source, size=8)
```

```python
show(u)
```
