```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
dados = pd.read_excel('/content/PLOT.xlsx')
```

```python
dadospx = px.data.gapminder()
```

```python
px.pie(dadospx[(dadospx.country == 'Spain')&(dadospx.continent == 'Europe')], values = 'year', names = 'year')
```

```python
px.scatter(dados, x  = 'A', y = 'B', color = 'C', marginal_x = 'histogram', marginal_y = 'violin')
```

```python
dados2 = dados.iloc[:50,:]
```

```python
dados2.plot.bar(figsize=(10,10),stacked = True)
```

```python
dados2.plot.bar(x='GRUPO',
                y='A', 
                figsize=(10,10))
```

```python
px.bar(dados2, x='CLASS', y='A')
```

```python
dados2.plot.line(figsize=(10,10))
```

```python
dados2.plot.line(x='CLASS', y = 'A', figsize = (10,10))
```

```python
px.line(dados2, x = range(0,50), y = 'A')
```
